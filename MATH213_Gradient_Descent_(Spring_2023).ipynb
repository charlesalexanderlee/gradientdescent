```python
'''
Charles Lee
Liam Nicholson
MATH213: Gradient Descent Explorations Extra Credit Project
Spring 2023
functions.py
'''

import numpy as np

'''
Problem 1:
'''
def f1(x, y):
    return x**2 + (y**2)/4

def grad_f1(x, y):
    return np.array([
        2*x,
        y/2
    ], dtype='float64')

'''
Problem 2:
- f2_1 is the first function defined in problem 2
- f2_2 is the second function (the Rosenbrock function) defined in problem 2
'''
def f2_1(x, y):
    return (x**2)*y*np.exp((-x**2)-(y**2))

def f2_2(x, y):
    return (1-x)**2 + 100*(y-x**2)**2

def grad_f2_1(x, y):
    return np.array([
        2*x*y*np.exp(-(x**2+y**2))*(1-x**2),
        (x**2)*np.exp(-(x**2+y**2))*(1-2*y**2)
    ], dtype='float64')

def grad_f2_2(x, y):
    return np.array([
        400*(x**3)+(2-400*y)*x-2,
        200*(y-x**2)
    ], dtype='float64')

'''
Problem 3:
- The proposed function defined here is a non-negative sinusoidal
  function decaying across the x-y plane, with the global minimum
  being 0 at f(0,0). The motivation behind this proposed function
  is that sinusoidal waves exhibit oscillatory behavior, therefore
  it will have multiple local minimums across the entire x-y plane.

  f(x, y) = [-sin(x)sin(y)/(xy)] + 1

'''
def f3(x, y):
    return -(np.sin(x)*np.sin(y))/(x*y) + 1

def grad_f3(x, y):
    return np.array([
        np.sin(y)*(np.sin(x)-x*np.cos(x)) / y*(x**2),
        1-((np.sin(x)*np.sin(y))/(x*y))
    ], dtype='float64')
```

```python
'''
Charles Lee
Liam Nicholson
MATH213: Gradient Descent Explorations Extra Credit Project
Spring 2023
graph.py
'''

import numpy as np
import matplotlib.pyplot as plt
# from functions import *

def plot(
    function: callable,
    gradient: callable,
    tau: float,
    num_iterations: int,
    seed_value: int,
    width: int,
    title: str
    ):
    # Define the range of x and y values
    x = np.linspace(-width, width, 50)
    y = np.linspace(-width, width, 50)

    # Create a grid of x and y values
    X, Y = np.meshgrid(x, y)

    # Evaluate the function at each point on the grid
    Z = function(X, Y)

    # Create a 3D plot
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')

    # Plot the surface function
    ax.plot_wireframe(X, Y, Z, alpha=0.25, color='gray')

    # For reproduciblity, we seed the random rnumber generator using the seed_value parameter
    np.random.seed(seed_value)

    # Plot 20 diverse paths for gradient descent
    for _ in range(20):
        # Initiliaze our current path with a random x and y coordinate
        curr_xy = np.array([
            np.random.randint(-width, width+1),
            np.random.randint(-width, width+1)
        ])

        # Assign a random color to each path
        color = np.random.rand(3,)

        # Plot the current path
        for _ in range(num_iterations):
            ax.scatter(curr_xy[0], curr_xy[1], function(curr_xy[0], curr_xy[1]), color=color)
            curr_xy = curr_xy - tau*gradient(curr_xy[0], curr_xy[1])

    # Add labels and title
    title_str = f'{title}\nTau={tau} | Iterations={num_iterations}'
    ax.set_title(title_str, weight='bold')
    ax.set_xlabel('x', weight='bold')
    ax.set_ylabel('y', weight='bold')
    ax.set_zlabel('f(x, y)', weight='bold')

    # Show the plot
    plt.show()
```
